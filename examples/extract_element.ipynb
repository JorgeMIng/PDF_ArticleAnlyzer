```python

i=3
acno = extract_element_soup(files[i],dir,element,type)
ners=[]
if len(acno)>0:
    for elements in acno:
        ners.extend(pipe(elements.text))



org_parser=False
misc_parser=False
org_temp=""
org_temp=dict()
org_temp["name"]=""
org_temp["score"]=0
orgs=[]
for ner in ners:
    print("name",ner["word"])
    if ner["entity"]=='B-ORG':
        org_parser=True
        org_temp["name"]=ner["word"]
        org_temp["score"]=ner["score"]
        print(ner["word"])
        continue
    if ner["entity"]=='B-MISC':
        misc_parser=True
        print("what")
        org_temp["name"]=ner["word"]
        org_temp["score"]=ner["score"]
        continue
    if (ner["entity"]=='I-ORG' or ner["entity"]=='I-MISC') and (org_parser or misc_parser):
        print("continue")
        org_temp["name"]=org_temp["name"]+ner["word"]
        org_temp["score"]=org_temp["score"]*ner["score"]
        continue
    if org_parser:
        print("shat")
        orgs.append(org_temp)
        org_temp["name"]=""
        org_temp["score"]=0
        org_parser=False
    if misc_parser:
        orgs.append(org_temp)
        org_temp["name"]=""
        org_temp["score"]=0
        misc_parser=False
    org_parser=False
    misc_parser=False
    
orgs    
```
